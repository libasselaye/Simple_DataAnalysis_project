# Suppress all the familiarisation Files

```python
df = pd.read_csv('participants.csv')
```

```python
# Path to the main directory containing participant folders
main_directory = '/content/drive/MyDrive/Python_Projet_Healthy/data'

# Iterate through each row of the DataFrame
for index, row in df.iterrows():
    participant_id = row['identifier']

    # Path to the participant's folder
    participant_directory = os.path.join(main_directory, str(participant_id))

    # Check if the participant's folder exists
    if os.path.exists(participant_directory):
        # List of files in the participant's folder
        files_in_directory = os.listdir(participant_directory)

        # Iterate through all files in the participant's folder
        for filename in files_in_directory:
            # Check if the file name starts with "familiarisation"
            if filename.startswith("familiarisation"):
                file_path = os.path.join(participant_directory, filename)

                # Delete the file
                os.remove(file_path)
                print(f"Le fichier {filename} a √©t√© supprim√©.")

    else:
        print(f"Dossier du participant {participant_id} introuvable.")
```

# Automatic Importation

```python
# Lire le fichier participants.csv
participants_df = pd.read_csv('participants.csv')
participants = participants_df['identifier'].to_list()
participants
```

```python
participants_df
```

```python
# Path to the main directory containing participant folders
main_directory = '/content/drive/MyDrive/Python_Projet_Healthy/data'
```

```python
#'    Var '
#'     IPe '
# Create an empty DataFrame to store data from dominant and non-dominant files
combined_data = []

# Iterate through each participant
for index, row in participants_df.iterrows():
    participant_id = row['identifier']

    # Dictionary to store participant data
    participant_data = {'identifier': [], 'VAR': [], 'Ipe': [], 'filesource': []}

    # Participant directory path
    participant_directory = os.path.join(main_directory, str(participant_id))

    # Check if participant directory exists
    if os.path.exists(participant_directory):
        # Iterate through files in participant directory
        for file in os.listdir(participant_directory):
            if file.endswith('.marker.csv'):
                # Full file path
                file_path = os.path.join(participant_directory, file)

                # Read the file
                file_df = pd.read_csv(file_path, on_bad_lines='skip', skiprows=17, delimiter=',')
                file_df.drop(file_df.index[:2], inplace=True)
                file_df = file_df.iloc[:,:-1]

                # Check if 'Var' and 'Ipe' columns exist in the file
                if '    Var ' in file_df.columns and '     IPe ' in file_df.columns:
                    # Add data to DataFrame
                    data = file_df[['    Var ', '     IPe ']]
                    data['filesource'] = file
                    data['identifier'] = participant_id
                    combined_data.append(data)
                else:
                    print(f"Columns 'Var' and 'Ipe' are not present in file {file}.")
    else:
        print(f"Participant directory {participant_id} not found.")

# Concatenate all data into a single DataFrame
combined_data_df = pd.concat(combined_data, ignore_index=True)

# Merge data with 'participants.csv' DataFrame
result_df = pd.merge(participants_df, combined_data_df, on='identifier', how='left')
```

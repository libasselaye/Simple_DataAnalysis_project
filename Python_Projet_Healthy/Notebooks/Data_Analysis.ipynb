# Add Column Main_Dominante

```python
#les données provenant des fichiers dominant1 et 2 vont avoir la valeur dominante
#pour cette colonne et les fichiers nondominante 1 et 2 seront affectées de non dominante
```

```python
dataset.head(5)
```

```python
dataset.filesource.unique()
```

```python
# Fonction pour appliquer la logique
def assign_main_dominante(row):
    if row['filesource'] == 'dominante1.marker.csv' or row['filesource'] == 'dominante2.marker.csv':
        return 'Dominante'
    else:
        return 'Non Dominante'
```

```python
# Appliquer la fonction à chaque ligne du DataFrame
dataset['Main_Dominante'] = dataset.apply(assign_main_dominante, axis=1)
dataset.head(10)
```

```python
# Define color palette for source files
file_colors = {'dominante1.marker.csv' : 'blue', 
               'dominante2.marker.csv' : 'green',
               'nondominante1.marker.csv' : 'red', 
               'nondominante2.marker.csv' : 'orange'}
```

```python
# function to plot the evolution curve for each participant
def plot_ipe_participant():
    participants = dataset['identifier'].unique()
    
    # Create a plot for each participant
    num_plots = len(participants)
    num_cols = 2
    num_rows = (num_plots + num_cols - 1) // num_cols
    
    fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, num_rows*5))
    axes = axes.flatten()
    
    for i, participant in enumerate(participants):
        ax = axes[i]
        for file, color in file_colors.items():
            ax.plot(dataset[(dataset['identifier'] == participant) & (dataset['filesource'] == file)]['Var'], 
                    dataset[(dataset['identifier'] == participant) & (dataset['filesource'] == file)]['IPe'],
                    label=file, color=color)
        ax.set_title(f'Courbe d\'évolution de Ipe pour {participant}')
        ax.set_xlabel('Enregistrement')
        ax.set_ylabel('Ipe')
        ax.grid(True)
        ax.tick_params(axis='x', rotation=45, labelsize=8)
        ax.legend()
    
    # Hide unused axis
    for j in range(i+1, len(axes)):
        axes[j].axis('off')
    
    plt.tight_layout()
    plt.show()

# Display evolution curves for each participant
plot_ipe_participant()
```
